We need to clean the following data: put it satandard and clean format, keeping only useful information.

```
df = pd.read_excel('Customer Call List.xlsx')
df
```

```
# Drop duplicates
df = df.drop_duplicates()
df
```

```
# Drop not useful column
df.drop(columns='Not_Useful_Column', inplace=True)
df
```

```
# Clean Last_Name column
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df.loc[:,'Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```
# Strip all numeric values to just numbers and format them uniformly
df.loc[:,'Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
df.loc[:,'Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df.loc[:,'Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df.loc[:,'Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df.loc[:,'Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

```
# Split address separated intro three different columns
split_columns = df['Address'].str.split(',', n=2, expand=True)
df[['Street_Address', 'State', 'Zip_Code']] = split_columns
df
```

```
# Replace Paying Customer and Do_Not_Contact with Y or N only
df.loc[:,'Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df.loc[:,'Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df.loc[:,'Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df.loc[:,'Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df
```

```
# Cleaning NaN and N/a
df = df.replace('N/a','')
df = df.fillna('')
df
```

```
# Keep only people with phone numbers and people who want to be contacted
df=df[df['Phone_Number']!='']
df=df[df['Do_Not_Contact']=='N']
df
```

```
df.reset_index(inplace=True, drop=True)
df
```

```
df.drop(columns='Address', inplace=True)
df
```
